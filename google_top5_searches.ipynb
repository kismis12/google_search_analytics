```python
import pandas as pd
import pytrends
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends=TrendReq()
```

```python
trends.trending_searches(pn='india')
```

```python
kw_list=["Tata Punch","Fabindia","IBPS PO notification 2021","Uttarakhand news","IRCTC share"]
trends.build_payload(kw_list,geo='IN')
data=trends.interest_by_region()
data.head(10)
print(data)
```

```python
lis=["Tata Punch","Fabindia","IBPS PO notification 2021","Uttarakhand news","IRCTC share"]
for i in lis:
    data=data.sort_values(by=i,ascending=False)
    data.reset_index().plot(x='geoName',y=i,figsize=(18,8),kind="bar")
    plt.style.use('fivethirtyeight')
    plt.show()
```

```python
kw_list=["Tata Punch","Fabindia","IBPS PO notification 2021","Uttarakhand news","IRCTC share"]
trends.build_payload(kw_list,timeframe= 'now 1-d', geo='IN')
data=trends.interest_by_region()
data.head(10)
print(data)
```

```python
lis=["Tata Punch","Fabindia","IBPS PO notification 2021","Uttarakhand news","IRCTC share"]
for i in lis:
    data=data.sort_values(by=i,ascending=False)
    data.reset_index().plot(x='geoName',y=i,figsize=(18,8),kind="bar")
    plt.style.use('fivethirtyeight')
    plt.show()
```

```python
data=TrendReq(hl='en-US' , tz=360)
kw_list=["Tata Punch","Fabindia","IBPS PO notification 2021","Uttarakhand news","IRCTC share"]
data.build_payload(kw_list,timeframe='now 1-d',geo='IN')
data=data.interest_over_time()
print(data)
```

```python
df=["Tata Punch","Fabindia","IBPS PO notification 2021","Uttarakhand news","IRCTC share"]
for i in df:
    fig,ax=plt.subplots(figsize=(18,8))
    data[i].plot()
    plt.style.use('fivethirtyeight')
    plt.xlabel('time')
    plt.ylabel('total count')
    plt.show()
    
```
